# **Проверка расчётов эффективности обогащения.**

```python
def calculate_recovery(df, C, F, T):
    C = df[C] # доля золота в концентрате после флотации
    F = df[F] # доля золота в сырье до флотации
    T = df[T] # доля золота в отвальных хвостах после флотации 
    
    return (C * (F - T)) / (F * (C - T)) * 100
recovery_check = calculate_recovery(train_data, 'rougher.output.concentrate_au', 'rougher.input.feed_au', 'rougher.output.tail_au')
mae = (train_data['rougher.output.recovery'] - recovery_check).abs().mean()
'MAE: {}'.format(mae)
```

Концентрация золота после флотации (C): C - это доля золота в образце концентрата после процесса флотации. Он представляет собой процентное содержание золота в полученном концентрате, который является продуктом флотационной обработки сырья.

Исходное содержание золота в сырье (F): F - это доля золота в исходном сырье до процесса флотации. Это значение представляет процентное содержание золота в исходном сырье перед его обработкой.

Остаточное содержание золота в отвальных хвостах после флотации (T): T - это доля золота в отвальных хвостах после флотации. Он представляет процентное содержание золота, которое остается в отвальных хвостах после извлечения золота в процессе флотации.

Оценка эффективности обогащения (Recovery): Эффективность обогащения рассчитывается с использованием значений C, F и T, и представляет собой процентное соотношение извлеченного золота относительно исходного содержания золота в сырье. Более высокие значения эффективности обогащения указывают на более эффективный процесс флотации и извлечения золота.

В данном случае, значение MAE равно 9.444891926100695e-15, что очень близко к нулю. Это указывает на то, что предсказанные значения эффективности обогащения практически идентичны фактическим значениям. Таким образом, можно считать, что данные о эффективности обогащения в таблице верны и функция recovery_check правильно рассчитывает эффективность обогащения.

# **Анализ изменения концентрации металлов (Au, Ag, Pb).**

```python
def plot_concentration(df, col1, col2, col3, col4):
    metals = ['au', 'ag', 'pb']
    names = {
        'au': 'золота',
        'ag': 'серебра',
        'pb': 'свинца'
    }
    for metal in metals:
        plt.figure(figsize=(12, 8), dpi=120)
        sns.kdeplot(df[col1 + metal], shade=True, color="purple", label="Исходная", alpha=.5)
        sns.kdeplot(df[col2 + metal], shade=True, color="orange", label="После флотации", alpha=.5)
        sns.kdeplot(df[col3 + metal], shade=True, color="blue", label="После первичной очистки", alpha=.5)
        sns.kdeplot(df[col4 + metal], shade=True, color="green", label="Финальная", alpha=.5)

        plt.title('Концентрация ' + names[metal] + ' на разных этапах', fontsize=16)
        plt.xlabel('Концентрация ' + names[metal])
        plt.ylabel('Плотность')
        plt.legend()
        plt.show()
```

```python
plot_concentration(full_data, 'rougher.input.feed_', 'rougher.output.concentrate_',
              'primary_cleaner.output.concentrate_', 'final.output.concentrate_')
```

После анализа графиков концентрации на разных этапах обработки мы можем сделать следующие выводы:

1. После флотации наблюдается повышение концентрации всех металлов. Это говорит о том, что процесс флотации успешно разделяет ценные металлы от примесей и позволяет увеличить содержание золота, серебра и свинца в концентрате.

2. На этапе первичной очистки особенно заметно повышение концентрации тяжелых металлов, таких как золото и свинец. Это указывает на эффективное удаление примесей и более глубокую отчистку от ценных металлов.

3. На финальной очистке наблюдается отделение золота от свинца. Концентрация золота значительно увеличивается, тогда как концентрация свинца остается на относительно низком уровне. Это свидетельствует о том, что финальная очистка успешно отделяет золото от свинца и позволяет получить концентрат с высокой содержащейся золота.

Таким образом, процессы флотации, первичной очистки и финальной очистки играют важную роль в повышении концентрации металлов и обогащении золотосодержащего сырья.

# **Сравнение распределений размеров гранул сырья.**

```python
def compare_grain_sizes(train_data, test_data):
    plt.figure(figsize=(12, 8), dpi=80)
    sns.distplot(train_data, color="purple", label="Train", hist_kws={'alpha': 0.7}, kde_kws={'linewidth': 3})
    sns.distplot(test_data, color="green", label="Test", hist_kws={'alpha': 0.4}, kde_kws={'linewidth': 3})
    plt.ylim(0, 0.06)

    plt.title('Распределение размеров гранул сырья', fontsize=22)
    plt.xlabel('Размер гранулы')
    plt.ylabel('Плотность')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.show()

compare_grain_sizes(train_data['rougher.input.feed_size'],
                    test_data['rougher.input.feed_size'].dropna())
```

Оба графика демонстрируют распределение размеров гранул сырья, которое имеет форму приближенную к гауссовскому распределению. Однако, можно заметить, что в тестовой выборке наблюдается небольшой сдвиг влево по сравнению с обучающей выборкой. Это означает, что в тестовой выборке преобладают более мелкие размеры гранул сырья по сравнению с обучающей выборкой.

Возможные причины этого сдвига могут быть различиями в процессе обработки или составе сырья, использованного в тестовой выборке. Небольшие различия в распределении размеров гранул могут потенциально влиять на производительность и эффективность процесса обогащения.

# **Исследование суммарных концентраций веществ.**

```python
def calculate_all_metal_concentration(df):
    columns = [
        'rougher.input.feed_au', 'rougher.input.feed_ag', 'rougher.input.feed_pb',
        'rougher.output.concentrate_au', 'rougher.output.concentrate_ag', 'rougher.output.concentrate_pb',
        'primary_cleaner.output.concentrate_au', 'primary_cleaner.output.concentrate_ag', 'primary_cleaner.output.concentrate_pb',
        'final.output.concentrate_au', 'final.output.concentrate_ag', 'final.output.concentrate_pb'
    ]

    df['rougher_input_feed_all_metal'] = df[columns[:3]].sum(axis=1)
    df['rougher_output_concentrate_all_metal'] = df[columns[3:6]].sum(axis=1)
    df['primary_output_concentrate_all_metal'] = df[columns[6:9]].sum(axis=1)
    df['final_output_concentrate_all_metal'] = df[columns[9:]].sum(axis=1)

    return df

def visualize_all_metal_concentration(df):
    plt.figure(figsize=(12, 8), dpi=100)
    colors = ["cyan", "magenta", "yellow", "green"]
    labels = ["Исходная", "После флотации", "После первичной очистки", "Финал"]

    for i, col in enumerate(
            ['rougher_input_feed_all_metal', 'rougher_output_concentrate_all_metal',
             'primary_output_concentrate_all_metal', 'final_output_concentrate_all_metal']):
        sns.kdeplot(df[col], shade=True, color=colors[i], label=labels[i], alpha=0.7)

    plt.title('Концентрация всех металлов на этапах очистки', fontsize=16)
    plt.legend()
    plt.show()

train_data = calculate_all_metal_concentration(train_data)
visualize_all_metal_concentration(train_data)
```

```python
def clean(df):
    before = df.shape[0]
    
    # Удаление строк с значениями меньше 1
    df = df.query('1 < rougher_input_feed_all_metal')
    df = df.query('1 < rougher_output_concentrate_all_metal')
    df = df.query('1 < primary_output_concentrate_all_metal')
    df = df.query('1 < final_output_concentrate_all_metal')
    
    after = df.shape[0]
    display('удалено {:.2%} строк'.format((before-after)/before))
    
    # Вычисление индексов строк с нулевой суммарной концентрацией
    zero_concentration_indices = df[df['rougher_input_feed_all_metal'] + df['rougher_output_concentrate_all_metal'] +
                                    df['primary_output_concentrate_all_metal'] + df['final_output_concentrate_all_metal'] == 0].index
    
    # Удаление строк с нулевой суммарной концентрацией 
    df.drop(zero_concentration_indices, inplace=True)
    
    return df



train_data = clean(train_data)
visualize_all_metal_concentration(train_data)
```

```python
def sMAPE(true, pred):
    """
    Рассчитывает симметричную среднюю абсолютную процентную ошибку (sMAPE) между истинными и предсказанными значениями.

    Параметры:
        true (массивоподобный): Истинные значения целевого признака.
        pred (массивоподобный): Предсказанные значения целевого признака.
        
    Возвращает:
        float: Значение sMAPE.
    
    Описание:
        Функция рассчитывает метрику sMAPE, которая измеряет процентное отклонение между истинными и предсказанными значениями.
        Она определяется как среднее абсолютное отклонение между двумя значениями, деленное на среднюю сумму их абсолютных значений,
        умноженное на 100. Полученное значение представляет собой среднюю процентную ошибку.
    """
    return (2 * np.abs(pred - true) / (np.abs(true) + np.abs(pred))).mean() * 100
```

```python
def total_sMAPE(sMAPE_rougher, sMAPE_final):
    """
    Рассчитывает финальное симметричное среднее абсолютное процентное отклонение (final sMAPE) 
    на основе значений sMAPE для обогащения после флотации и финального обогащения.
    
    Параметры:
        sMAPE_rougher (float): Значение sMAPE для обогащения после флотации.
        sMAPE_final (float): Значение sMAPE для финального обогащения.
        
    Возвращает:
        float: Значение final sMAPE.
    
    Описание:
        Функция рассчитывает финальное sMAPE, используя взвешенную сумму значений sMAPE для обогащения после флотации
        и финального обогащения. Коэффициенты взвешивания 0.25 и 0.75 соответственно задаются для учета важности каждой 
        метрики. Финальное sMAPE представляет собой среднюю процентную ошибку, учитывающую оба этапа процесса обогащения.
    """
    return (0.25 * sMAPE_rougher + 0.75 * sMAPE_final)
```

```python
def preprocess_test_data(df):
    """
    Предобрабатывает тестовые данные для модели.

    Параметры:
    - df: DataFrame, исходные тестовые данные.

    Возвращает:
    - df: DataFrame, предобработанные тестовые данные.

    """
    before = df.shape[0]

    # Добавляем целевые признаки из полного набора данных
    df = df.merge(full_data[['rougher.output.recovery', 'final.output.recovery']],
                  how='left', on='date')

    

    # Добавляем сумму металлов на входе
    df['rougher_input_feed_all_metal'] = (df['rougher.input.feed_au'] +
                                          df['rougher.input.feed_ag'] +
                                          df['rougher.input.feed_pb'])

    # Удаляем строки с низкими значениями и убираем выбросы
    df = df.query('1 < rougher_input_feed_all_metal')
    rougher_input_feed_all_metal_low = df['rougher_input_feed_all_metal'].quantile(0.025)
    rougher_input_feed_all_metal_up = df['rougher_input_feed_all_metal'].quantile(0.975)
    df = df.query('@rougher_input_feed_all_metal_low < rougher_input_feed_all_metal < @rougher_input_feed_all_metal_up')

    after = df.shape[0]
    display('Удалено {:.2%} строк'.format((before - after) / before))
    return df

test_data = preprocess_test_data(test_data)
```
